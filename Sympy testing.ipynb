```python
from sympy.solvers import solve
from sympy import Symbol
from sympy.solvers.inequalities import solve_univariate_inequality
from sympy import *
```

```python
x = Symbol('x')
```

```python
func = (x**2 - 2*x + 1)/x
```

```python
solve_univariate_inequality(func >= 0, x)
```

```python
simplify(func)
```

```python
t = Symbol('t')
expfunc = func.subs(x, exp(t))
```

```python
solve_univariate_inequality(expfunc < 0, t)
```

```python
p1 = plot(func, xlim = (-1, 1), ylim = (-10, 10))
p2 = plot(expfunc, xlim = (-1, 1), ylim = (-3, 3))
p1
p2
```

```python
y = Symbol('y')
poly2 = Poly(y**3 - 6.001 * y**2 + 11 * y - 6, y)
solve_poly_inequality( poly2, '>=')
```

```python
init_printing()
solutions = solveset(poly2, domain=Reals)
```
