```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x, y):
    z = x**2 + y**2
    return z
```

```python
x = np.arange(-30,30)
y = np.arange(-30,30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(X, Y, Z)
```
