```python
import numpy as np
from astropy import units as u
from astropy import uncertainty as unc
np.random.seed(12345)  # ensures reproducible example numbers
```

```python
c = unc.normal(1*u.kpc, std=30*u.pc, n_samples=1000000)
d = unc.uniform(center=3*u.kpc, width=800*u.pc, n_samples=1000000)
e = unc.Distribution(((np.random.beta(2,5, 1000000)-(2/7))/2 + 3)*u.kpc)
f = (c * d * e) ** (1/3)
# f.pdf_mean() 
# f.pdf_std() 
from matplotlib import pyplot as plt 
from astropy.visualization import quantity_support 
with quantity_support():
     plt.hist(f.distribution, bins=50)
```
