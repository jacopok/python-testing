```python
import numpy.random as rand
import numpy as np
```

```python
N = int(1e4)
```

```python
class astronaut:
    thr = .95

    def __init__(self):
        self.skill = rand.uniform(high=thr)
        self.luck = rand.uniform(high=1-thr)
        self.score = self.skill + self.luck
        
```

```python
a = [(i, astronaut()) for i in range(N)]
```
